### Univariate Analysis - Categorical Variables

```python
# Frequency Distribution and Percentage for Categorical Variables
categorical_columns = ['Area', 'City', 'Type of Dining']
frequency_distributions = {}
for var in categorical_columns:
    freq_dist = dataset[var].value_counts().reset_index()
    freq_dist.columns = [var, 'Count']
    freq_dist['Percentage'] = ((freq_dist['Count'] / freq_dist['Count'].sum())*100).round(1)
    frequency_distributions[var] = freq_dist
```

4. Walk in Offer and Rating: The correlation coefficient is -0.06, indicating a negligible negative correlation, meaning there is no meaningful relationship between the walk-in offer and the rating.
5. Walk in Offer and Average distance from City-Center: The correlation coefficient is 0.03, which is a negligible positive correlation, implying no meaningful relationship between the walk-in offer and the distance from the city center.
6. Rating and Average distance from City-Center: The correlation coefficient is 0.06, indicating a negligible positive correlation, suggesting no significant relationship between restaurant rating and its distance from the city center.
